```
data=pd.read_csv("n_with_connection.csv")
plt.figure(figsize=(15,10))
plt.plot(data['N'],data['Connection'],"ro-",label="actual data")
x=np.linspace(0,data['N'].max(),100)
y=x**1.135
plt.plot(x,y,"b",label="y=x**1.135")
y2=x*8
plt.plot(x,y2,"y",label="y=8x")
plt.xlabel("N")
plt.ylabel("Connections")
plt.legend()
```
