```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("StepsAndDistance.csv")
```

## 1. Guess the function according to graph

```python
plt.figure(figsize=(15,10))
x=np.array(data['steps'],dtype=float)
plt.plot(data['steps'],data['distance'],'-',label="random walk simulation")
y=x**0.48
y2=x**0.5
plt.ylabel("distance")
plt.xlabel("steps")
plt.plot(x,y,'pr-',label="y=x^0.48")
plt.plot(x,y2,'py-',label="y=x^0.5")
plt.legend()
```

## 2. Get the power parameter of the minimal difference

```python
def getDifference(y,y1):
    return abs((y1-y).sum())
def getSquarXDifference(parameters):
    x=np.array(data['steps'],dtype=float)
    y=np.array(data['distance'],dtype=float)
    difflist=list()
    for i in parameters:
        y1=x**i
        diff=getDifference(y1,y)
        difflist.append(diff)
    return np.array(difflist)
```

```python
x=np.linspace(0.48,0.5,100)
y=getSquarXDifference(x)
```

According to following graph when the power parameter is arround at 0.49, the gussed function has the minimal difference with simulation test.

```python
plt.plot(x,y,'r-')
```

Calculate the best parameter 

```python
x[y.argmin()]
```

## 3. Draw the guessed line again

The relationship between steps(n) and distance(d) is d=n^0.4903

```python
plt.figure(figsize=(15,10))
x=np.array(data['steps'],dtype=float)
plt.plot(data['steps'],data['distance'],'-',label="Random walk simulation")
y=x**0.4903
yupper=x**0.47
ylower=x**0.51
plt.ylabel("distance")
plt.xlabel("steps")
plt.plot(x,y,'r-',label="y=x^0.4903")
plt.plot(x,yupper,'b-',label="y=x^0.47")
plt.plot(x,ylower,'y-',label="y=x^0.51")
plt.legend()
```
